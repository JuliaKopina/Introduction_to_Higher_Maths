1. Напишите код на Python, реализующий построение графика окружности

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_circle(R=1,x0=0,y0=0): # на вход подаётся радиус, координаты центра в точке x0,y0 
    x1 = np.linspace(-R, R, 100)
    x2 = np.linspace(R,-R,100)
    y = np.append(np.sqrt((R**2-x1**2)), -np.sqrt((R**2-x2**2)))
    x = np.append(x1,x2, axis=0)
    graph_border = R*1.25
    plt.xlim(-graph_border+x0, graph_border+x0)
    plt.ylim(-graph_border+y0, graph_border+y0)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(b=None, which='major', axis='both')
    plt.plot(x+x0,y+y0)
```

```python
draw_circle(3,1,1)
```

2. Напишите код на Python, реализующий построение графика эллипса

```python
def draw_ellipse(a=3,b=2,x0=0,y0=0): # на вход подаётся a, b, x0,y0 - центр графика
    t = np.linspace(0, 2*np.pi, 100)
    x_border = a*1.25
    y_border = b*1.25
    x = a*np.cos(t)
    y = b*np.sin(t)
    plt.xlim(-x_border+x0, x_border+x0)
    plt.ylim(-y_border+y0, y_border+y0)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(b=None, which='major', axis='both')
    plt.plot(x+x0,y+y0)
```

```python
draw_ellipse(4,2)
```

```python
def draw_ellipse2(a=4,b=2,x0=0,y0=0): # на вход подаётся a, b, x0,y0 - центр графика
    x1 = np.linspace(-a, a, 100)
    y1 = (b/a)*np.sqrt(a**2-x1**2)
    x2 = np.linspace(a,-a,100)
    y = np.append(y1, -y1)
    x = np.append(x1,x2, axis=0)
    x_border = a*1.25
    y_border = b*1.25
    plt.xlim(-x_border+x0, x_border+x0)
    plt.ylim(-y_border+y0, y_border+y0)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.grid(b=None, which='major', axis='both')
    plt.plot(x+x0,y+y0)
```

```python
draw_ellipse2(4,2)
```
